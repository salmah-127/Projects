From this data, we can observe that there are 642 instances of individuals at high risk of developing heart disease within ten years, compared to 3,595 instances of individuals not at risk. This indicates a significant imbalance in the dataset. In practical terms, identifying individuals at high risk of heart disease is more critical than identifying those not at risk. Therefore, it is necessary to balance the data categories to prevent subsequent models from being biased towards the majority class.

```python
data = pd.concat([X, y], axis=1)

majority_class = data[data['TenYearCHD'] == 0]
minority_class = data[data['TenYearCHD'] == 1]

majority_downsampled = resample(majority_class,
                                replace=False,
                                n_samples=len(minority_class),
                                random_state=42)

downsampled_data = pd.concat([majority_downsampled, minority_class])
downsampled_value_counts = downsampled_data['TenYearCHD'].value_counts()

plt.figure(figsize=(8, 4))
downsampled_value_counts.plot(kind='bar', color=['blue', 'orange'])
plt.title('Downsampled Value Counts of TenYearCHD')
plt.xlabel('TenYearCHD')
plt.ylabel('Counts')
plt.xticks(rotation=0)
plt.show()
```

We employed an undersampling method to randomly select data from class 0, making the number of samples equal to that of class 1.

##Feature Selection

Since We have 15 features, and we obtain scores for the relationship between features and the target variable using SelectKBest. The higher the score, the stronger the correlation. We then sort the features in descending order based on their scores and select the features with scores greater than 0.

```python
# Create SelectKBest feature selector, using mutual information
X_downsampled = downsampled_data.drop(columns=['TenYearCHD'])
y_downsampled = downsampled_data['TenYearCHD']

selector = SelectKBest(score_func=lambda X, y: mutual_info_classif(X, y, random_state=42), k=15)
X_new = selector.fit_transform(X_downsampled, y_downsampled)

print("Original number of features:", X.shape[1])
print("Reduced number of features:", X_new.shape[1])

scores = selector.scores_
feature_scores = pd.DataFrame({'Feature': X.columns, 'Score': scores,})
feature_scores = feature_scores.sort_values(by='Score', ascending=False)
print(feature_scores)
```

```python
# Sort data for better visualization
feature_scores = feature_scores.sort_values(by='Score', ascending=False)

plt.figure(figsize=(7, 5))
sns.barplot(x='Score', y='Feature', data=feature_scores, palette='viridis')
plt.xlabel('Score')
plt.ylabel('Feature')
plt.title('Feature Importance')
plt.show()
```

##Tune parameters

The aforementioned algorithms were utilized with default parameters. Upon evaluating the algorithms, it was observed that there is improvement required for the expected performance. Consequently, the next step involves tuning the parameters of these three algorithms to further enhance their F1-score. Finally, the algorithms will be tested on a separate test dataset to evaluate their performance on general data.

```python
# Function to find the best parameters for a model
def find_best_parameters(model, param_grid):
  # Create GridSearchCV
  grid_search = GridSearchCV(model, param_grid, scoring='f1', cv=kf, verbose=1, n_jobs=-1, refit='F1')
  grid_search.fit(X_train, y_train)
  best_f1_score = grid_search.best_score_

  # Print the best parameters and best F1-score
  print("Best parameters:", grid_search.best_params_)
  print("Best cross-validation F1-score: {:.4f}%".format(grid_search.best_score_*100))
  return grid_search.best_estimator_, best_f1_score
```

## Accuracy, Precision, Recall, and F1-Score

Confusion matrix is used to evaluate the performance of each model, the label position of confusion matrix in python is shown as followed:

| Column 1| Column 2 |
|---------|----------|
| TN      | FP       |
| FN      | TP       |

Accuracy measures the overall correctness of the classifier.

Precision measures the proportion of true positive predictions among all positive predictions made by the classifier.

Recall measures the proportion of true positive predictions that were correctly identified by the classifier out of all actual positives.

F1 score is the harmonic mean of precision and recall, providing a balance between the two metrics.

```python
# Function to evaluate the model on the test data
def evaluate_test_model(model, X_test, y_test):
    y_pred = model.predict(X_test)

    # Evaluate the model performance
    cm = confusion_matrix(y_test, y_pred)
    accuracy = accuracy_score(y_test, y_pred) * 100
    precision = precision_score(y_test, y_pred) * 100
    recall = recall_score(y_test, y_pred) * 100
    f1 = f1_score(y_test, y_pred) * 100
    print(f"\nModel: {model.__class__.__name__}")
    print('\nConfusion Matrix:')
    print(cm)
    print('\nTest Set Accuracy: {:.4f}%'.format(accuracy))
    print('Test Set Precision: {:.4f}%'.format(precision))
    print('Test Set Recall: {:.4f}%'.format(recall))
    print('Test Set F1-Score: {:.4f}%'.format(f1))

    return accuracy, precision, recall, f1
```

```python
# Initialize lists to store evaluation metrics
models = [lr_best_model, dt_best_model, svm_best_model, rf_best_model, ada_best_model, gbm_best_model]
model_names = ['Logistic Regression', 'Decision Tree', 'SVM', 'Random Forest','Adaboost','Gradient Boosting Machine']
accuracy_values = []
precision_values = []
recall_values = []
f1_values = []

# Evaluate models and store metrics
for model, name in zip(models, model_names):
    metrics = evaluate_test_model(model, X_test, y_test)
    accuracy_values.append(metrics[0])
    precision_values.append(metrics[1])
    recall_values.append(metrics[2])
    f1_values.append(metrics[3])
```

## ROC Curve and PR curve

```python
# Plot ROC curves
for model, name in zip(models, model_names):
  y_pred_proba = model.predict_proba(X_test)[:, 1] if hasattr(model, "predict_proba") else model.decision_function(X_test)
  fpr, tpr, _ = roc_curve(y_test, y_pred_proba)
  roc_auc = auc(fpr, tpr)
  plt.plot(fpr, tpr, lw=2, label=f'{name} (area = {roc_auc:.2f})')

plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")

plt.show()
```

AUC-ROC is used to evaluate how well a binary classification model can distinguish between two classes. The closer the curve to the top-left corner, the higher the TPR and low FPR across thresholds.

Among the six classification models, **SVM and Adaboost** are both strong performers with area under the curve greater than 0.70. The Logistic Regression curve is the lowest, indicating less effective performance in distinguishing between classes.

```python
# Plot Precision-Recall curves
for model, name in zip(models, model_names):
  y_pred_proba = model.predict_proba(X_test)[:, 1] if hasattr(model, "predict_proba") else model.decision_function(X_test)
  precision, recall, _ = precision_recall_curve(y_test, y_pred_proba)
  pr_auc = auc(recall, precision)
  plt.plot(recall, precision, lw=2, label=f'{name} (area = {pr_auc:.2f})')

plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.legend(loc="lower left")

# plt.tight_layout()
plt.show()
```

The precision-recall curve shows the tradeoff between precision and recall for different threshold. A high area under the curve represents both high recall and high precision, where high precision relates to a low false positive rate, and high recall relates to a low false negative rate.

Based on the PR curve, **SVM and Adaboost** have better performance in balancing their precision and recall scores (area under the curve 0.7 and 0.68 respectively). Meanwhile, Logistic Regression has the lowest area under the curve with only 0.60.
