```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
from tqdm import tqdm

# Load model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Fill NaNs with empty strings for now to avoid issues
df['symptoms'] = df['symptoms'].fillna('')
df['clinical_note'] = df['clinical_note'].fillna('')

# Generate embeddings
tqdm.pandas()
df['symptoms_emb'] = df['symptoms'].progress_apply(lambda x: model.encode(x))
df['clinical_notes_emb'] = df['clinical_note'].progress_apply(lambda x: model.encode(x))
```

```python
#Text imputation function

def impute_column(df, target_col, emb_col):
    df_imputed = df.copy()

    # Identify rows to impute
    missing_mask = df_imputed[target_col].str.strip() == ''
    complete_rows = df_imputed[~missing_mask]

    for idx in df_imputed[missing_mask].index:
        query_vec = df_imputed.loc[idx, emb_col]

        # Compare to all rows with non-missing text
        similarities = cosine_similarity(
            [query_vec],
            list(complete_rows[emb_col])
        )[0]

        best_match_idx = similarities.argmax()
        imputed_value = complete_rows.iloc[best_match_idx][target_col]

        df_imputed.at[idx, target_col] = imputed_value

    return df_imputed

# Impute symptoms
df = impute_column(df, target_col='symptoms', emb_col='clinical_notes_emb')

# Impute clinical_notes
df = impute_column(df, target_col='clinical_note', emb_col='symptoms_emb')


df.head(5)
```

##Sentiment Analysis

To create a text input for sentiment analysis using Mistral, the symptoms and clinical_note columns were combined into a single combined_text column. This combined text was then processed in batches, where each entry was classified as either positive or negative based on the sentiment model. The sentiment labels were mapped back to the original dataset by linking the symptoms and clinical_note combinations. The sentiment distribution was calculated and visualized using a pie chart, and a summary of the sentiment percentages was fed into Mistral to generate an interpretation of the overall sentiment patterns.

```python
import torch
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, AutoModelForCausalLM
from accelerate import Accelerator

# Load Mistral-7B Model on GPU

mistral_model_name = "mistralai/Mistral-7B-Instruct-v0.1"
mistral_tokenizer = AutoTokenizer.from_pretrained(mistral_model_name)
mistral_tokenizer.pad_token = mistral_tokenizer.eos_token
mistral_model = AutoModelForCausalLM.from_pretrained(mistral_model_name, torch_dtype=torch.float16)

# Initialize Accelerator for managing GPU usage
accelerator = Accelerator()

# Prepare Mistral model for GPU usage using Accelerator
mistral_model, mistral_tokenizer = accelerator.prepare(mistral_model, mistral_tokenizer)

```

```python
df_copy = df.copy()


# Step 1: Get unique combinations of symptoms + clinical_note

unique_df = df_copy[['symptoms', 'clinical_note']].drop_duplicates().copy().reset_index(drop=True)

# Combine both fields for sentiment input
unique_df['combined_text'] = unique_df.apply(
    lambda row: f"Symptoms: {row['symptoms']}\nClinical Note: {row['clinical_note']}", axis=1
)


# Step 2: Sentiment function using Mistral (binary: positive/negative)

def get_sentiment_from_mistral(texts, batch_size=4):
    sentiments = []
    for i in range(0, len(texts), batch_size):
        batch = texts[i:i+batch_size]
        prompts = [
            f"Classify the sentiment of this patient statement as either positive or negative:\n\n\"{txt}\"\n\nSentiment:"
            for txt in batch
        ]

        inputs = mistral_tokenizer(prompts, return_tensors="pt", padding=True, truncation=True).to(mistral_model.device)

        with torch.no_grad():
            outputs = mistral_model.generate(
                **inputs,
                max_new_tokens=10,
                do_sample=False,
                pad_token_id=mistral_tokenizer.eos_token_id
            )

        decoded = mistral_tokenizer.batch_decode(outputs, skip_special_tokens=True)

        for full_response in decoded:
            last_line = full_response.strip().split("\n")[-1].lower()
            if "positive" in last_line:
                sentiments.append("positive")
            else:
                sentiments.append("negative")
    return sentiments


# Step 3: Apply sentiment analysis
unique_df['sentiment'] = get_sentiment_from_mistral(unique_df['combined_text'].tolist())

# -----------------------------
# Step 4: Map sentiments back to the original df
# Create a key from the combo of symptoms + clinical_note
df['sentiment'] = df.set_index(['symptoms', 'clinical_note']).index.map(
    dict(zip(unique_df.set_index(['symptoms', 'clinical_note']).index, unique_df['sentiment']))
)

df.head(10)
```

```python
# Calculate sentiment distribution
sentiment_counts = df['sentiment'].value_counts(normalize=True) * 100

# Pie chart
plt.figure(figsize=(6, 6))
plt.pie(
    sentiment_counts.values,
    labels=sentiment_counts.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=['#66bb6a', '#ef5350']  # green for positive, red for negative
)

plt.title("Sentiment Distribution", fontsize=14)
plt.axis('equal')
plt.show()
```

```python
#Sentiment analysis intrepretation with Mistral
# 1. Prepare the sentiment summary text
summary_text = "\n".join([
    f"{label}: {value:.2f}%" for label, value in sentiment_counts.items()
])

# 2. Create the prompt for Mistral
interpretation_prompt = (
    "Given the following sentiment analysis distribution from patient symptoms and clinical notes:\n\n"
    f"{summary_text}\n\n"
    "Please provide a concise interpretation of the clinical sentiment pattern observed in this dataset."
)

# 3. Tokenize and generate the interpretation with Mistral
inputs = mistral_tokenizer(interpretation_prompt, return_tensors="pt").to(mistral_model.device)

with torch.no_grad():
    output = mistral_model.generate(
        **inputs,
        max_new_tokens=100,
        do_sample=False,
        pad_token_id=mistral_tokenizer.eos_token_id
    )

# 4. Decode and print the interpretation
interpretation = mistral_tokenizer.decode(output[0], skip_special_tokens=True)
print("\nInterpretation of Sentiment Distribution:\n")
print(interpretation.strip())
```
